```python
sales = [
    {"product": "Laptop", "amount": 15000},
    {"product": "Mouse", "amount": 500},
    {"product": "Keyboard", "amount": 1000},
    {"product": "Laptop", "amount": 15000},
    {"product": "Mouse", "amount": 500}
]
```

```python
total_income = sum(item["amount"] for item in sales)
print("Toplam Gelir:", total_income)
```

```python
from collections import Counter

product_count = Counter(item["product"] for item in sales)

print("Ürün Satış Adetleri:")
print(product_count)
```

```python
product_income = {}

for item in sales:
    product = item["product"]
    amount = item["amount"]

    if product in product_income:
        product_income[product] += amount
    else:
        product_income[product] = amount

print("Ürün Bazlı Gelir:")
print(product_income)
```

```python
most_sold = product_count.most_common(1)[0]
print("En çok satılan ürün:", most_sold)
```
